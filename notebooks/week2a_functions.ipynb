1. **–æ–ø—Ä–µ–¥–µ–ª—è–µ–º** –æ–¥–∏–Ω —Ä–∞–∑ —Å –ø–æ–º–æ—â—å—é `def`,

def introduce():      # def + –∏–º—è + —Å–∫–æ–±–∫–∏ + –¥–≤–æ–µ—Ç–æ—á–∏–µ

## 3. –ü–∞—Ä–∞–º–µ—Ç—Ä—ã, –∫–æ—Ñ–µ–º–∞—à–∏–Ω–∞ –∏ –≥–ª–æ–±–∞–ª—å–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è (—Å–ª–∞–π–¥—ã 4‚Äì6)

–ù–∞ —Å–ª–∞–π–¥–∞—Ö –±—ã–ª–∞ –∞–Ω–∞–ª–æ–≥–∏—è:

- —Ñ—É–Ω–∫—Ü–∏—è ‚Äî —ç—Ç–æ **–∫–æ—Ñ–µ–º–∞—à–∏–Ω–∞** ‚òï,
- –ø–∞—Ä–∞–º–µ—Ç—Ä ‚Äî —ç—Ç–æ **—Ç–∏–ø –∫–æ—Ñ–µ** (‚Äú–ª–∞—Ç—Ç–µ‚Äù, ‚Äú—ç—Å–ø—Ä–µ—Å—Å–æ‚Äù, ...).

–ü—Ä–∏–º–µ—Ä —Å –§—É–Ω—Ç–∏–∫–æ–º:

- –≥–ª–æ–±–∞–ª—å–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è `total_food` —Ö—Ä–∞–Ω–∏—Ç, **—Å–∫–æ–ª—å–∫–æ –≤—Å–µ–≥–æ –æ–Ω —Å—ä–µ–ª –∑–∞ –¥–µ–Ω—å**;
- —Ñ—É–Ω–∫—Ü–∏—è `add_food(amount)`:
  - –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä `amount` (—Å–∫–æ–ª—å–∫–æ –≥—Ä–∞–º–º–æ–≤ –∫–æ—Ç —Å—ä–µ–ª –ø—Ä—è–º–æ —Å–µ–π—á–∞—Å),
  - –¥–æ–±–∞–≤–ª—è–µ—Ç —ç—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫ `total_food`.

–í–∞–∂–Ω–æ –ø–æ–Ω—è—Ç—å:

- **–ø–∞—Ä–∞–º–µ—Ç—Ä** ‚Äî —ç—Ç–æ *–º–µ—Å—Ç–æ –≤ —Ñ—É–Ω–∫—Ü–∏–∏*, –∫—É–¥–∞ –ø–æ–¥—Å—Ç–∞–≤–ª—è–µ—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ;
- **–∞—Ä–≥—É–º–µ–Ω—Ç** ‚Äî —ç—Ç–æ *–∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ*, –∫–æ—Ç–æ—Ä–æ–µ –º—ã –ø–µ—Ä–µ–¥–∞—ë–º –ø—Ä–∏ –≤—ã–∑–æ–≤–µ.

````python
# üí∞ –§—É–Ω—Ç–∏–∫ —Å—á–∏—Ç–∞–µ—Ç, —Å–∫–æ–ª—å–∫–æ –æ–Ω —Å—ä–µ–ª –∑–∞ –¥–µ–Ω—å

# –ì–ª–æ–±–∞–ª—å–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è –Ω–∞–∫–æ–ø–ª–µ–Ω–∏—è –µ–¥—ã
total_food = 0

def add_food(amount):
    """
    –£–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç –≥–ª–æ–±–∞–ª—å–Ω—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é total_food
    –Ω–∞ amount –≥—Ä–∞–º–º–æ–≤ –∏ –ø–µ—á–∞—Ç–∞–µ—Ç –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç.
    """
    global total_food
    total_food = total_food + amount
    print(f"üê± –§—É–Ω—Ç–∏–∫ —Å—ä–µ–ª {amount} –≥. –í—Å–µ–≥–æ: {total_food} –≥")

# –ó–∞–≤—Ç—Ä–∞–∫
add_food(50)

# –û–±–µ–¥
add_food(70)

# –£–∂–∏–Ω
add_food(60)

print(f"\nüìä –ò—Ç–æ–≥–æ –∑–∞ –¥–µ–Ω—å: {total_food} –≥—Ä–∞–º–º–æ–≤")
````

## 5. –§—É–Ω–∫—Ü–∏—è —Å –¥–≤—É–º—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏: –∫–æ—Ç –ë–∞–π—Ç–∏–∫ –∏ –µ–≥–æ –±–∏—Ç–∫–æ–π–Ω‚Äë–∫–æ—à–µ–ª—ë–∫ (—Å–ª–∞–π–¥ 8)

–ü—Ä–∏–º–µ—Ä —Å–æ —Å–ª–∞–π–¥–∞:

- –≥–ª–æ–±–∞–ª—å–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è `balance` —Ö—Ä–∞–Ω–∏—Ç –±–∞–ª–∞–Ω—Å –≤ –±–∏—Ç–∫–æ–π–Ω–∞—Ö (‚Çø);
- —Ñ—É–Ω–∫—Ü–∏—è `transaction(amount, name)`:
  - –ø—Ä–∏–Ω–∏–º–∞–µ—Ç **–¥–≤–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞**:
    - `amount` ‚Äî —Å–∫–æ–ª—å–∫–æ –ø—Ä–∏—à–ª–æ –∏–ª–∏ —É—à–ª–æ,
    - `name` ‚Äî –Ω–∞–∑–≤–∞–Ω–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏;
  - –≤–Ω—É—Ç—Ä–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç `if`, —á—Ç–æ–±—ã –∫—Ä–∞—Å–∏–≤–æ –æ—Ñ–æ—Ä–º–∏—Ç—å –¥–æ—Ö–æ–¥ –∏–ª–∏ —Ä–∞—Å—Ö–æ–¥.

–°–µ–π—á–∞—Å –ø–æ–≤—Ç–æ—Ä–∏–º —ç—Ç–æ—Ç –ø—Ä–∏–º–µ—Ä –∏ —á—É—Ç—å –µ–≥–æ —Ä–∞—Å—à–∏—Ä–∏–º.

````python
# üíª –ö–æ—Ç –ë–∞–π—Ç–∏–∫ –ø—Ä–æ–≤–µ—Ä—è–µ—Ç —Å–≤–æ–π —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω—ã–π –∫–æ—à–µ–ª—ë–∫

balance = 0  # –≥–ª–æ–±–∞–ª—å–Ω—ã–π –±–∞–ª–∞–Ω—Å –≤ –±–∏—Ç–∫–æ–π–Ω–∞—Ö

def transaction(amount, name):
    """
    –ú–µ–Ω—è–µ—Ç –≥–ª–æ–±–∞–ª—å–Ω—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é balance –Ω–∞ amount
    –∏ –ø–µ—á–∞—Ç–∞–µ—Ç —Å—Ç—Ä–æ–∫—É –æ —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–∏.
    """
    global balance
    balance = balance + amount
    if amount > 0:
        print(f"‚úÖ {name}: +{amount}‚Çø | –ë–∞–ª–∞–Ω—Å: {balance}‚Çø")
    else:
        print(f"‚ùå {name}: {amount}‚Çø | –ë–∞–ª–∞–Ω—Å: {balance}‚Çø")

# –ü–æ—Å—Ç—É–ø–ª–µ–Ω–∏—è –∏ —Ç—Ä–∞—Ç—ã
transaction(100, "–ó–∞—Ä–ø–ª–∞—Ç–∞")
transaction(-30, "–ö–æ—Ä–º")
transaction(-20, "–ò–≥—Ä—É—à–∫–∏")
transaction(50, "–ü—Ä–µ–º–∏—è")

print(f"\nüí∞ –ò—Ç–æ–≥–æ–≤—ã–π –±–∞–ª–∞–Ω—Å: {balance}‚Çø")
````

## 6. –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–∞ `global`: –≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∏–∑ —Ñ—É–Ω–∫—Ü–∏–∏

–í –ø—Ä–∏–º–µ—Ä–∞—Ö –≤—ã—à–µ –º—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏ `global`, —á—Ç–æ–±—ã –º–µ–Ω—è—Ç—å –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é `balance` –≤–æ –≤—Å–µ–π –ø—Ä–æ–≥—Ä–∞–º–º–µ.
–ò–Ω–æ–≥–¥–∞ —É–¥–æ–±–Ω–µ–µ, –∫–æ–≥–¥–∞ —Ñ—É–Ω–∫—Ü–∏—è **–≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –Ω–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ**, –∞ –º—ã —Å–∞–º–∏ –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º –µ–≥–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.

–°–¥–µ–ª–∞–µ–º –Ω–µ–±–æ–ª—å—à—É—é –≤–µ—Ä—Å–∏—é `transaction_no_global`, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ–±–Ω–æ–≤–ª—ë–Ω–Ω—ã–π –±–∞–ª–∞–Ω—Å.

````python
def transaction_no_global(balance, amount, name):
    """
    –ù–µ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç global: –ø–æ–ª—É—á–∞–µ—Ç –±–∞–ª–∞–Ω—Å –∏ —Å—É–º–º—É,
    –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –Ω–æ–≤—ã–π –±–∞–ª–∞–Ω—Å.
    """
    new_balance = balance + amount
    if amount > 0:
        print(f"‚úÖ {name}: +{amount}‚Çø | –ë–∞–ª–∞–Ω—Å: {new_balance}‚Çø")
    else:
        print(f"‚ùå {name}: {amount}‚Çø | –ë–∞–ª–∞–Ω—Å: {new_balance}‚Çø")
    return new_balance

balance2 = 0
balance2 = transaction_no_global(balance2, 100, "–ó–∞—Ä–ø–ª–∞—Ç–∞")
balance2 = transaction_no_global(balance2, -30, "–ö–æ—Ä–º")
balance2 = transaction_no_global(balance2, -20, "–ò–≥—Ä—É—à–∫–∏")
balance2 = transaction_no_global(balance2, 50, "–ü—Ä–µ–º–∏—è")

print(f"\nüí∞ –ò—Ç–æ–≥–æ–≤—ã–π –±–∞–ª–∞–Ω—Å (–±–µ–∑ global): {balance2}‚Çø")
````

## 7. –ö–æ—Ç‚Äë—à–ø–∏–æ–Ω –®—Ç–∏—Ä–ª–∏—Ü: —Å–µ–∫—Ä–µ—Ç–Ω—ã–π –¥–Ω–µ–≤–Ω–∏–∫ –∏ —Ñ—É–Ω–∫—Ü–∏—è `cat_encode` (—Å–ª–∞–π–¥ 9)

````python
# üïµ –ö–æ—Ç –®—Ç–∏—Ä–ª–∏—Ü –≤–µ–¥—ë—Ç —Å–µ–∫—Ä–µ—Ç–Ω—ã–π –¥–Ω–µ–≤–Ω–∏–∫

def cat_encode(text):
    """
    –ó–∞–º–µ–Ω—è–µ—Ç –≤–∞–∂–Ω—ã–µ —Å–ª–æ–≤–∞ –Ω–∞ '–∫–æ—à–∞—á–∏–π –∫–æ–¥'.
    """
    code = {"–≤—Å—Ç—Ä–µ—á–∞": "–º—è—É-–º—è—É", "–≤—Ä–∞–≥": "–≥–∞–≤-–≥–∞–≤", "—É—Å–ø–µ—Ö": "–º—É—Ä-–º—É—Ä"}
    for word, cat_word in code.items():
        text = text.replace(word, cat_word)
    return text

report = "–°–µ–≥–æ–¥–Ω—è –±—ã–ª–∞ –≤—Å—Ç—Ä–µ—á–∞. –í—Ä–∞–≥ –Ω–µ –∑–∞–º–µ—Ç–∏–ª. –£—Å–ø–µ—Ö!"

# –ó–∞–ø–∏—Å—ã–≤–∞–µ–º –∑–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω—ã–π —Ç–µ–∫—Å—Ç –≤ —Ñ–∞–π–ª
with open("secret_diary.txt", "w", encoding="utf-8") as f:
    f.write(cat_encode(report))

# –ß–∏—Ç–∞–µ–º —Ç–µ–∫—Å—Ç –æ–±—Ä–∞—Ç–Ω–æ –∏–∑ —Ñ–∞–π–ª–∞
with open("secret_diary.txt", "r", encoding="utf-8") as f:
    print("üîí –°–µ–∫—Ä–µ—Ç–Ω—ã–π –¥–Ω–µ–≤–Ω–∏–∫:")
    print(f.read())

# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ: –∑–∞–∫–æ–¥–∏—Ä—É–µ–º –µ—â—ë –æ–¥–Ω–æ —Å–æ–æ–±—â–µ–Ω–∏–µ –±–µ–∑ –∑–∞–ø–∏—Å–∏ –≤ —Ñ–∞–π–ª
extra_report = "–ù–æ–≤–∞—è –≤—Å—Ç—Ä–µ—á–∞ –∑–∞–≤—Ç—Ä–∞. –£—Å–ø–µ—Ö –≥–∞—Ä–∞–Ω—Ç–∏—Ä–æ–≤–∞–Ω."
print("\nüìù –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –∑–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ:")
print(cat_encode(extra_report))
````
